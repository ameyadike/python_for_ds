```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x,y,'r')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented method

fig = plt.figure()

axis = fig.add_axes([0.1,0.1,0.8,0.8])

axis.plot(x,y)
axis.set_xlabel('X Label')
axis.set_ylabel('Y label')
axis.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Large plot')

axes2.plot(y,x,'r')
axes2.set_title('Smaller plot')
```
